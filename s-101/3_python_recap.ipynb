### Conditions

```python
# since we can't always control what is in our variables (e.g user input or unknown effects) we may need to perform conditional actions
# this is where the if-then-else construct comes into play

#a = 3
#a = "error"
if type(a) == int: #<- this is called the conditional
  print(a+1)
else :
  print("not an integer")
```

```python
# we can also add more complex logic to the conditional
# and, or, not <-> &, |, !

a = 3.2
#a = 3
#a = "error"
if (type(a) == int) or (type(a) == float):
  print(a+1)
else :
  print("not an integer")
```

### All together

```python
# side effects: we often use loops to manipulate objects

typelist = []
mylist = [5.0, "guitar", 10, True]

# the append is a method of lists whose job is to add the element to the end
for element in mylist:
  typelist.append(type(element))

typelist

#the advantage here is that we may still use typelist for further processing, whereas before we just printed the types and lost access to them
```

```python
# nested loops
cool_list = [[1,2,3,4],["eenie", "meenie", "miny", "moe"],[True,False,True]]

for element in cool_list:
  for sub_element in element:
    print(sub_element)
```

### While Loops

```python
# With the while loop we can execute a set of statements as long as a condition is true.

i = 0

while i < 6: # statement --> here the statment is true if the variable i is smaller that 6
  i += 1
  print(i)
```

```python
# continue in while loops 

i = 0

while i < 6: # statemnet 
  i += 1
  if i == 3: # condition 
    continue # we stop the current iteration and continue with the next one 
  else:
    print(i)
```

```python
# break 

i = 0

while i < 6: # statment 
  i += 1
  if i == 3: # condition 
    break # with the break we can stop the loop even if the while condition is true:
  print(i)

print('hello world ')
```
